```python
from sys import stdin

def spiralPrint(mat, nRows, mCols):
    #Your code goes here
    cs = 0
    ce = mCols - 1

    rs = 0
    re = nRows - 1

    i=0
    j=0

    while( cs<=ce and rs<=re):

        i = rs
        j = cs


        while(j <= ce):
            print(mat[i][j], end = " ")
            j += 1

        rs += 1

        i = rs
        j = ce

        while(i <= re):
            print(mat[i][j], end = " ")
            i += 1

        ce -= 1

        i = re
        j = ce

        while(j >= cs):
            print(mat[i][j], end = " ")
            j -= 1

        re -= 1

        i = re
        j = cs

        while(i >= rs):
            print(mat[i][j], end = " ")
            i -= 1

        cs += 1

        

    































#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


```

```python
#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    spiralPrint(mat, nRows, mCols)
    print()

    t -= 1
```
